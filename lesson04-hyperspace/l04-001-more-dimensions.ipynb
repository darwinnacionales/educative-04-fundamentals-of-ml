```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns

# data
x1, x2, x3, y = np.loadtxt("pizza_3_vars.txt", skiprows=1, unpack=True)
w = np.array([-3.98230894, 0.37333539, 1.69202346])

# style
sns.set(rc={"axes.facecolor": "white", "figure.facecolor": "white"})

# figure
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection="3d")
ax.set_xlabel("Temperature", labelpad=15, fontsize=15)
ax.set_ylabel("Reservations", labelpad=15, fontsize=15)
ax.set_zlabel("Pizzas", labelpad=25, fontsize=15)
ax.zaxis.set_label_coords(1.03, 0.5)

# points
ax.scatter3D(x1, x2, y, color="b")

# plane
MARGIN = 10
edges_x = [x1.min() - MARGIN, x1.max() + MARGIN]
edges_y = [x2.min() - MARGIN, x2.max() + MARGIN]
xs, ys = np.meshgrid(edges_x, edges_y)
zs = w[0] + xs * w[1] + ys * w[2]
ax.plot_surface(xs, ys, zs, alpha=0.2)

fig.subplots_adjust(left=0.20, bottom=0.15, right=0.87, top=0.95)
plt.show()
```
